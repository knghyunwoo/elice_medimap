```python
import json

with open('num_of_doctors_for_detail_page.json') as file:
    json_string = file.read()    
    ambulance = json.loads(json_string)

final = []
for i in ambulance:
    temp = {
        "type" : i["type"],
        "name" : i["name"],
        "시" : i["metro"],
        "도" : i["suburb"],
        "전체 평균" : i["total_avg"],
        "시 평균" : i["metro_avg"],
        "도 평균" : i["suburb_avg"]
    }
    final.append(temp)
    
print(final)
```

```python
# 상세페이지 json 형태로 만들어주는 부분
with open('num_of_doctors_for_detail_page_changed.json', 'w', encoding='UTF-8') as file:
    jsonified = json.dumps(final)
    file.write(jsonified)
    
```
